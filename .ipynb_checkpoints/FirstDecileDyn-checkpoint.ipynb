```python
import math as m
import matplotlib as mpl
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
from numpy import inf
from scipy import optimize
from scipy.interpolate import make_interp_spline
from datetime import datetime
```

```python
def ModelVars(W): #if inputting W arrayed in time, use transpose/Wplot (axis 0 groups, axis 1 time)
    #calculate variables in the ODE
    P = m_P*np.tanh(W/W_0) + P_0 #a hypothetical production function scaling with wealth to an asymptote
    S = eps*P #salary income by group
    sumINV = (1-eps)*np.sum(P,axis=0) #available investment income
    Wpolynom = a0 + a1*W + a2*W**2 #polynomial determining share in investment income by wealth. W**1 for simple interest, W**2 for scaling interest, W**0 for "welfare"
    r_in = sumINV/np.sum(Wpolynom,axis=0) #rate of return on investment given relation between available income and wealth. sum over groups if arrayed in time
    INV = np.dot(Wpolynom,r_in) #investment income by group. dot product to also enable multiplication with W and r_in arrayed in time
    D = D_0 + d*W #dissipation of wealth by group
    return (P,S,r_in,INV,D)

def Diffusion(W): #find the component of dW/dt due to random mobility/diffusion
    
    #define relevant fields
    L = (W[1:]+W[:-1])/2 #define midpoints between wealth group averages
    rho = 1/ngroups/(W[1:]-W[:-1]) #approximate average density of individuals at midpoints
    L = np.insert(L, 0, 2*W[0]-L[0]) #insert additional off-points for L grid
    L = np.insert(L, ngroups, 2*W[-1]-L[-1])
    rho = np.insert(rho, 0, rho[0]/2) #define densities at off-points as about half of densities at adjacent in-points
    rho = np.insert(rho, ngroups, rho[-1]/2)
    
    #spline rho to uniform grid
    rhospl = make_interp_spline(L,rho) #splining function
    W_min = L[0] #find limits of splining space
    W_max = L[-1]
    gridpoints = (len(L))*10 + 1 #number of points in grid
    Wgrid = np.linspace(W_min,W_max,gridpoints) #wealth grid
    DelW = Wgrid[1] - Wgrid[0] #grid spacing
    rhogrid = rhospl(Wgrid) #splined rho values on wealth grid
    
    #find mobility function (M) and derivative (dMdW) on grid
    M = b0 + b1*Wgrid
    dMdW = b1*np.ones(gridpoints)
    
    #solve diffusion equation on grid for Delrho
    d0 = M[1:-1]*(rhogrid[2:]-2*rhogrid[1:-1]+rhogrid[:-2])/DelW**2 #regular diffusion
    d1 = dMdW[1:-1]*(rhogrid[2:]-rhogrid[:-2])/2/DelW #varying diffusion constant
    drhodtgrid = d0 + d1 #total
    
    #interpolate drhodt back to L grid
    splback = make_interp_spline(Wgrid[1:-1],drhodtgrid) #spline result
    drhodt = splback(L) #use spline to put back on L grid
    
    #solve dWdt based on Delrho assuming the change in the lowest group is 0
    dWdt = np.zeros(ngroups) #initialize dWdt, we will keep dWdt[0] = 0 for now
    dWdt[1] = - rho[1]**(-2)/ngroups*drhodt[1] 
    for i in range(2,ngroups):
        dWdt[i] = dWdt[i] - rho[i]**(-2)/ngroups*drhodt[i] #this is not even entirely correct, does not account for the change in L_i's
    dWdtavg = np.mean(dWdt) #find the average movement of groups
    dWdt -= dWdtavg #adjust values by setting the net movement due to diffusion to 0
    
    return dWdt

def AltDiffusion(W): #find the component of dW/dt due to random mobility/diffusion using a "diffusion velocity"
    
    #define relevant fields
    L = (W[1:]+W[:-1])/2 #define midpoints between wealth group averages
    rho = 1/ngroups/(W[1:]-W[:-1]) #approximate average density of individuals at midpoints
    rho[rho==inf] = 10 #set infinite to fixed value
    
    L = np.insert(L, 0, 2*W[0]-L[0]) #insert additional off-points for L grid
    L = np.insert(L, ngroups, 2*W[-1]-L[-1])
    rho = np.insert(rho, 0, rho[0]/2) #approximate densities at off-points as half of densities at adjacent in-points
    rho = np.insert(rho, ngroups, rho[-1]/2)
    
    delrho = (rho[1:]-rho[:-1])/(L[1:]-L[:-1]) #approximate density gradients at the streamline W's
    delrho[L[1:]-L[:-1]==0] = 0 #set divide by zero to fixed value
    rhoW = rho[:-1]+delrho*(W-L[:-1]) #use these to better approximate densities at the W's
    M = b0 + b1*W #calculate mobility at the various grid points given base (b0) and linearly increasing component (b1)
    
    #compute diffusion velocity
    diffvel = - M*delrho/rhoW
    diffvel -= np.mean(diffvel) #regularize by subtracting mean (diffusion should not result in net movement)
    
    return diffvel
    
def NextTimeStepWealth(W,dt): #the ODE
    #call function to calculate ODE variables
    P,S,r_in,INV,D = ModelVars(W)
    
    #call function to calculate diffusion component
    DIF = AltDiffusion(W)
    
    #find wealth at next timestep
    deltaW = dt*(S+INV-D+DIF) #change in wealth
    W_new = W + deltaW #new wealth
    W_new = np.heaviside(W_new,0)*W_new #harder limit at 0
    
    return W_new
```

```python
def Run(ic, ngroups, dt, time): #function to generate a solution over time
    ntimesteps = int(time/dt) #find number of timesteps
    
    W = np.zeros((ntimesteps,ngroups)) #generate a 2-D array to store W at all timesteps
    W[0] = ic #put in provided initial conditions
    
    for i in range(1,ntimesteps): #run the ODE
        W[i] = NextTimeStepWealth(W[i-1],dt)
        if i%100000 == 0: #output progress every 5000 timesteps
            print('timestep: ' + str(i))
            
    return W
```

```python
#Parameters

#production parameters (in tanh function)
m_P = 1 #production scaling
W_0 = 30 #wealth scaling parameter inside tanh
P_0 = 0.1 #base production

#investment parameters
eps = 0.5 #proportion of net production given to investors (as opposed to producers)
a0 = 50 #scalar for "welfare" in investment share polynomial
a1 = 1 #scalar for simple interest
a2 = 0 #0.01 0.02 #scalar for scaling interest

#diffusion parameters
b0 = 3 #scalar component of mobility
b1 = b0/100 #slope of linear component of mobility

#used in dissipation linear function
D_0 = 0.15 #base consumption
d = 0.01 #depreciation scalar

ngroups = 50 #number of income groups
```

```python
#Generate sample initial conditions
W_init = np.zeros(ngroups)
for i in range(ngroups):
    W_init[i] = (i+1)*200/ngroups
```

```python
#plot results

logplot = False #set if plotting log or absolute

W_plot = np.transpose(Result) #generate wealth vs. time arrays by trasponsing result

fig = plt.figure(figsize = (13,13))
ax = fig.add_subplot(111)
ntimesteps = int(runtime/timestep)
t = np.linspace(0, ntimesteps*timestep, ntimesteps)

#initialize colormap
colornorm = cm.colors.Normalize(vmin = 0, vmax = ngroups-1) #normalize color range to list of groups
scalarcolormap = cm.ScalarMappable(norm=colornorm,cmap='hsv') #create the scalar map using that norm

#plot decile wealth
if logplot:
    for i in range(ngroups): #plot group by group
        color = scalarcolormap.to_rgba(i) #use prev. defined scalarcolormap to fetch color as four dim RBGA-vector
        ax.plot(t, np.log(W_plot[i]+1), c = color) #, label = 'Decile ' + str(i+1)
    #plot average wealth
    W_average = np.mean(W_plot,axis=0)
    ax.plot(t, np.log(W_average+1), 'k-', label = 'Average')
    ax.set_ylabel('Log(Wealth+1)', fontsize = 14) #set y-label for log plot
else:
    for i in range(ngroups): #plot group by group
        color = scalarcolormap.to_rgba(i) #use prev. defined scalarcolormap to fetch color as four dim RBGA-vector
        ax.plot(t, W_plot[i], c = color) #, label = 'Decile ' + str(i+1)
    #plot average wealth
    W_average = np.mean(W_plot,axis=0)
    ax.plot(t, W_average, 'k-', label = 'Average')
    ax.set_ylabel('Wealth', fontsize = 14) #set y-label for absolute plot

#set plot labels, etc
ax.set_title('Wealth over Time By Decile', fontsize = 20)
ax.set_xlabel('Time', fontsize = 14)

#find percentage holding 50% of total wealth at end
halfwealth = W_average[ntimesteps-1]*ngroups/2
for i in range(1,ngroups):
    halfwealth -= Result[ntimesteps-1][-i]
    if halfwealth <= 0:
        percentfor50 = 100*(i - abs(halfwealth)/Result[ntimesteps-1][-i])/ngroups
        break
        
#find percentage at 0 wealth
percentat0 = 0
for i in range(ngroups):
    if Result[ntimesteps-1][i] >= 0.1: #find first group above 0 wealth
        percentat0 = i/ngroups*100 #calculate percentage using number of groups below that
        break

#add a dummy plotline to insert average wealth at end, % holding 50%, and % at 0 as legend text, then plot legend
ax.plot([],[], ' ', label='Average wealth: ' + str(round(W_average[ntimesteps-1],1)) + '\n' 
        + '% holding 50% of wealth: ' + str(round(percentfor50,2)) + '%\n' + '% at 0 wealth: ' + str(round(percentat0,2)) + '%')
ax.legend(fontsize = 12, loc = 2)

#save figure
save = True #whether or not to save !!!SET!!!
if save:
    np.savetxt('Images/Same as Absolute Plot.txt', [0]) #save a dummy text file explaining the plot
    
    now = datetime.now() #get current date and time
    dt_string = now.strftime("%Y-%m-%d %H-%M-%S") #convert to file-nameable string
    plt.savefig('Images/' + dt_string + '.png') #save figure with date and time
```
